```python
from bs4 import BeautifulSoup
import requests
from time import sleep

# ある2chのスレを遡っていくスクリプト
# 記事を紹介しているレスを表示していく


def main():
    # 前スレのurlを識別
    shikibestu = "net/test/read.cgi/future"

    # 特定の2chスレ
    url = "https://rio2016.2ch.net/test/read.cgi/future/1506697885/"

    # 3ページまで
    for d in range(3):

        # ページ解析
        HTML = requests.get(url)
        soup = BeautifulSoup(HTML.content, "lxml")
        c = soup.find("div", class_="message")

        for i in c:
            k = i.find_all("a")

            for n in k:
                # リンク先を識別、表示
                if shikibestu in n.text:
                    print(n.text)

                    # urlのあるレスを引っ張ってくる
                    kuro = soup.find_all("div", class_="post")
                    for aaa in kuro:
                        if "http" in aaa.text or "https" in aaa.text:
                            print(aaa.text)
                            print("-"*120)
                            print()  # 改行


        # 次の探索先
        url = n.text
        sleep(3)


if __name__ == "__main__":
    main()
        
```
